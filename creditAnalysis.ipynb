# Build Predictive Model
Predict whether an applicant's loan will be approved given the applicant characteristics.
Potential things to consider:
- Which characteristics to use
- What models to try

```python
# Simple example
from sklearn.linear_model import LogisticRegression #for the model
from sklearn.metrics import accuracy_score #for scoring metrics

predictor_cols = ['unemployed', 'gender', 'marriageStatus', 'problematicRegion', 'age', 'accountAmount', 'loanAmount', 'payoffMonths', 'yrsAtCompany']
model = LogisticRegression().fit(df_example[predictor_cols], df_example['approved'])
df_example['prediction'] = model.predict(df_example[predictor_cols])

accuracy_score(df_example['approved'], df_example['prediction'])
```
